```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

games = pd.read_csv('/content/drive/Shareddrives/302 Project/Data/GameSales.csv')

ax = sns.boxplot(data=games, x='genre', y='critic_score', order=['Action', 'Adventure', 'Action-Adventure', 'Fighting', 'Misc', 'MMO', 'Platform', 'Puzzle', 'Racing', 'Role-Playing', 'Shooter', 'Simulation', 'Sports', 'Strategy'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = 90)
ax.set_xlabel("Genre", fontsize = 20)
ax.set_ylabel("Critic Scores", fontsize = 20)

```
